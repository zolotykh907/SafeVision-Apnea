```python
import xml.etree.ElementTree as ET
```

```python
def get_attrubuts(root, space, namespace, tag, family, tip):
    '''
    root - корень файла xml
    space - где надо искать аттрибуты
    namespace - пространство имен, где все искать
    tag - на что должен заканчиваться тег в пространстве имен
    '''

    data = root.find(f'ns:{space}', {'ns': namespace})
    if data is None:
        return []

    target_arr = []

    events = None
    for child in data:
        if child.tag.endswith(tag):
            events = child
            break

    if events is None:
        return []

    for event in events:
        event_attributs = event.attrib
        if event_attributs['Family'] == family and event_attributs['Type'] == tip:
            target_arr.append(event_attributs)

    return target_arr
```

```python
tree = ET.parse('/var/data/apnea/rml/00000999-100507.rml')
root = tree.getroot()

namespace = 'http://www.respironics.com/PatientStudy.xsd'
space = 'ScoringData'
tag = "Events"

apnea_attributs = get_attrubuts(root, space, namespace, tag, 'Respiratory', 'ObstructiveApnea')
```
